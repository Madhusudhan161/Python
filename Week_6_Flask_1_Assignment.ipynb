### 1 . What is Flask framework and its Advantages ?

##### Flask :
- flask is a micro web framework written in python. it is classsified as mircroweb framework, because it does not require any tool or libraries. it has no database abstraction layer, form validation, or any other components where pre-existing third party libraries pprovide common functions.

- it is designed to create and scale web apps quickly an simply.

##### Advantages of Flask :

1. It is a lightweight framework that offers hassle free development.
2. Provides flexibility to developers to experiment with thier modules or architecture.
3. It is suitable for small projects .
4. Offers a built in development server and fast debugger.
5. Easily scalable for applications.
6. Support for secure cookies

### 3 . What is app routing in Flask ? why do we use it ?

##### App routing : 

- App routing means mapping URL to a specific function that will handle the logic for that URL. Thre latest framework use the routing technique to help users remember application URL. 


##### Why do we use App routing :
- The route () decorator in flask is used to bind URL to a function. as a result when a URL is mentioned in the browser, the function is executed to give the result.
